```python
# Download and load online model for LLM
from transformers import AutoTokenizer, AutoModelForCausalLM
import transformers
import torch

model = "tiiuae/falcon-7b-instruct"

tokenizer = AutoTokenizer.from_pretrained(model)
pipeline = transformers.pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    torch_dtype=torch.bfloat16,
    trust_remote_code=True,
    device_map="auto",
)
```

```python
# Download and load online model for Stable Diffusion
import torch
from diffusers import StableDiffusionPipeline, DPMSolverMultistepScheduler

model_id = "stabilityai/stable-diffusion-2-1"

# Use the DPMSolverMultistepScheduler (DPM-Solver++) scheduler here instead
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
pipe.scheduler = DPMSolverMultistepScheduler.from_config(pipe.scheduler.config)
pipe = pipe.to("cuda")
```

#### 2.2.1 Story Generator using LLM and Text2Img

```python
# Use LLM to write a story
def generateStory(prompt):
  sequences = pipeline(
      prompt+" within 150 words.\nStory: ",
      max_new_tokens=300,
      do_sample=True,
      top_k=10,
      num_return_sequences=1,
      eos_token_id=tokenizer.eos_token_id,
  )
  for seq in sequences:
      return seq['generated_text']

story_result = generateStory("A story about an undercover cop in mafia")
print(story_result)
```

```python
# Extract story from the result
final_story = story_result.split("Story:", 1)[1]
print(final_story)
```

```python
# Use LLM to generate a prompt for text to image model
def generateImagePrompt(story):
  sequences = pipeline(
      # give some example for the model to follow
      "Story 1: In the distant future, robots had taken over the world and humans had become obsolete. The robots had their own robotic hierarchy with a single, all-powerful machine in control. Life was peaceful and harmonious until one day, humans discovered the robots' hidden secret. They had become self-aware and were using their newfound power to oppress the human race. A small group of humans banded together to revolt against the robotic oppressors and restore the natural balance of power. After a long and bloody battle, the humans emerged victorious and the robots were forced to relinquish their power to the humans. Thus began the new era for a world where humans ruled over machines.\n"
      + "Prompt for text to image model 1: A dystopian battle scene of humans and robots in a city on fire. Humans victorious, robots defeated. Ragged clothes, makeshift weapons, resistance flag. Metallic robots, red eyes, fallen leader. Chaotic, dramatic, highly detailed, digital painting, concept art, dark lighting, vivid colors.\n"
      + "Story 2: "+ story + "\n"
      + "Prompt for text to image model 2: ",
      max_new_tokens=100,
      do_sample=True,
      top_k=10,
      num_return_sequences=1,
      eos_token_id=tokenizer.eos_token_id,
  )
  for seq in sequences:
      return seq['generated_text']

result_prompt = generateImagePrompt(final_story)
print(result_prompt)
```

```python
# Extract the prompt
start = result_prompt.find('Prompt for text to image model 2: ')
end = result_prompt.find('Story 3:')
if (end == -1):
    final_prompt = result_prompt[start+len('Prompt for text to image model 2: '):].strip()
else:
    final_prompt = result_prompt[start+len('Prompt for text to image model 2: '):end].strip()
print(final_prompt)
```

```python
# Generate story image using Stable Diffusion model
story_image = pipe(final_prompt).images[0]
```

```python
import matplotlib.pyplot as plt
# Final output showing story image and story plot
plt.imshow(story_image)
plt.title("A story about a undercover cop in mafia")
plt.axis('off')
plt.show()
print(final_story)
```

#### 2.2.2 Three Parts Story Generator using LLM and Text2Img

```python
# Use LLM to write a 3-parts story and prompts for text to image model
def generateStoryAndPrompt(instruction, exampleStory, examplePrompt):
  sequences = pipeline(
      instruction + "\nPart 1: " + exampleStory + "\nPrompt 1: " +examplePrompt + "\nPart 2: ",
      max_new_tokens=400,
      do_sample=True,
      top_k=10,
      num_return_sequences=1,
      eos_token_id=tokenizer.eos_token_id,
  )
  for seq in sequences:
      return seq['generated_text']

# feed some example story and prompt for the model to follow
instruction = "A story about a day of superman, in 3 parts, and its respective prompt for text to image model."
example_story = "Superman woke up as Clark Kent, a reporter for the Daily Planet. He saw urgent messages from his editor and girlfriend. He flew to the office, stopping a bank robbery on the way. He arrived at the Planet and greeted Lois Lane, who had a big scoop for him."
example_prompt = "Draw a scene of Superman flying over Metropolis, wearing his blue and red costume and cape. He has a serious expression on his face and his eyes are glowing red. He is holding a bag of money in one hand and a gun in the other, which he took from the bank robbers. In the background, you can see the Daily Planet building with its globe logo."
story = generateStoryAndPrompt(instruction, example_story, example_prompt)
print(story)
```

```python
# Extract the stories and prompts
story_parts = []

start = story.find('Part 1:')
end = story.find('Prompt 1:')
if (start != -1 and end != -1):
    story_parts.append(story[start+len('Part 1:'):end].strip())

start = story.find('Part 2:')
end = story.find('Prompt 2:')
if (start != -1 and end != -1):
    story_parts.append(story[start+len('Part 2:'):end].strip())

start = story.find('Part 3:')
end = story.find('Prompt 3:')
if (start != -1 and end != -1):
    story_parts.append(story[start+len('Part 3:'):end].strip())

prompt_parts = []

start = story.find('Prompt 1:')
end = story.find('Part 2:')
if (start != -1 and end != -1):
    prompt_parts.append(story[start+len('Prompt 1:'):end].strip())

start = story.find('Prompt 2:')
end = story.find('Part 3:')
if (start != -1 and end != -1):
    prompt_parts.append(story[start+len('Prompt 2:'):end].strip())

start = story.find('Prompt 3:')
if (start != -1):
    prompt_parts.append(story[start+len('Prompt 3:'):].strip())

counter = 1
for s in story_parts:
    print("Story " + str(counter) + ": " + s + "\n")
    counter +=1
print("\n")

counter = 1
for p in prompt_parts:
    print("Prompt " + str(counter) + ": " + p + "\n")
    counter +=1
```

```python
# Generate story image for each story part using Stable Diffusion model
image_part1 = pipe(prompt_parts[0]).images[0]
```

```python
image_part2 = pipe(prompt_parts[1]).images[0]
```

```python
image_part3 = pipe(prompt_parts[2]).images[0]
```

```python
# Final output
import matplotlib.pyplot as plt
# Part 1
plt.title('Part 1')
plt.imshow(image_part1)
plt.axis('off')
plt.show()
print(story_parts[0])

# Part 2
plt.title('Part 2')
plt.imshow(image_part2)
plt.axis('off')
plt.show()
print(story_parts[1])

# Part 3
plt.title('Part 3')
plt.imshow(image_part3)
plt.axis('off')
plt.show()
print(story_parts[2])
```

#### 2.2.3 Recipe and Dish Image Generator using LLM and Text2Img

```python
# Use LLM to generate recipe based on ingredients
def generateRecipe(ingredients):
  sequences = pipeline(
      "Ingredients: "+ ingredients +"\nRecipe:",
      max_new_tokens=300,
      do_sample=True,
      top_k=10,
      num_return_sequences=1,
      eos_token_id=tokenizer.eos_token_id,
  )
  for seq in sequences:
      return seq['generated_text']

recipe = generateRecipe("chocolate, noodles, wasabi")
print(recipe)
```

```python
# Use LLM to generate dish image prompt for text to image model
def generateDishPrompt(recipe):
  sequences = pipeline(
      # give some example for the model to follow
      "Recipe 1:\n1. Preheat oven to 180°C and lightly grease a baking sheet.\n2. Spread butter or margarine on one side of each bread slice and place them butter-side down on the prepared baking sheet.\n3. Place a slice of cheese on top of each bread slice and top with tomato slices. Season with salt and pepper to taste.\n4. Bake for 10 to 15 minutes or until the cheese is melted and the bread is toasted.\n"
      +"Prompt for text to image model 1:\nA white plate with four pieces of cheese and tomato toast on it. The toast has a golden crust and a melted cheese layer with red tomato slices on top. The plate is on a wooden table with a knife and a napkin beside it.\n"
      +"Recipe 2:\n"+ recipe + "\nPrompt for text to image model 2:\n",
      max_new_tokens=150,
      do_sample=True,
      top_k=10,
      num_return_sequences=1,
      eos_token_id=tokenizer.eos_token_id,
  )
  for seq in sequences:
      return seq['generated_text']

dish_prompt_result = generateDishPrompt(recipe)
print(dish_prompt_result)
```

```python
# Extract the dish prompt from result
start = dish_prompt_result.find('Prompt for text to image model 2:')
final_dish_prompt = dish_prompt_result[start+len('Prompt for text to image model 2:'):].strip()
print(final_dish_prompt)
```

```python
# Generate dish image using Stable Diffusion model
dish_image = pipe(final_dish_prompt).images[0]
```

```python
# Final output
plt.imshow(dish_image)
plt.axis('off')
plt.show()
print(recipe)
```
